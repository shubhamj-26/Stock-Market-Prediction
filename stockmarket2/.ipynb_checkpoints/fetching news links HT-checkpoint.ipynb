```python
#Python program to scrape website  
#and save quotes from website 
import requests 
from bs4 import BeautifulSoup 
import csv 
import re

URL = "https://www.hindustantimes.com/topic/mumbai"
r = requests.get(URL) 
#print(r)
  
soup = BeautifulSoup(r.content, 'html.parser') 
#print(soup)
quotes=[]  # a list to store quotes 
  
table = soup.find_all('div', attrs = {'class':'para-txt'}) 
#print(table)

for ik in table:
    links = re.findall("href=[\"\'](.*?)[\"\']", str(ik))
    print(links[0])

```

```python
#Python program to scrape website  
#and save quotes from website 
import requests 
from bs4 import BeautifulSoup 
import csv 
import re

def callingnews():

    URL = "https://www.usatoday.com/search/?q=amazon"
    r = requests.get(URL) 
#print(r)
  
    soup = BeautifulSoup(r.content, 'html.parser') 
#print(soup)
    quotes=[]  # a list to store quotes 
  

    table1 = soup.find_all('a', attrs = {'class':'gnt_se_a gnt_se_a__hd gnt_se_a__hi'}) 
#print(table1)

#table13 = table11.get_text()
#print(table13) 

    table11 = soup.find_all('div', attrs = {'class':'gnt_pr'}) 

    datalist=[]
    linksdata=[]
#print(table11)
    for ik in table11:
        datalist.append(ik.get_text())
        print(ik.get_text())

    pos=0
    listtocheck=[]
    for ik in table1:
        links = re.findall("href=[\"\'](.*?)[\"\']", str(ik))
        linksdata.append(links[0])
        if 'story' not in links[0]:
            listtocheck.append(pos)
        pos+=1
        print(links)

    print(listtocheck)

    for ij in range(len(listtocheck)):
        print(ij)
        datalist.pop(ij)
        linksdata.pop(ij)
    #print(listtocheck[ij])

    print(len(datalist))
    print(len(linksdata))
    return datalist,linksdata
```
